```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the seaborn style
sns.set(style="whitegrid")

# Load the dataset into a DataFrame
df = pd.read_csv('/content/Cardiotocographic.csv')

# Display the first few rows of the dataset
df.head()

# 1. Data Cleaning and Preparation

# Check for missing values
print("Missing values in each column:\n", df.isnull().sum())

# Handle missing values (e.g., imputation or deletion)
# For simplicity, let's drop rows with missing values
df_clean = df.dropna()

# Identify and correct data type inconsistencies
print("\nData types:\n", df_clean.dtypes)

# Convert any columns that should be numeric but are stored as strings
# Example: df_clean['column_name'] = pd.to_numeric(df_clean['column_name'], errors='coerce')

# Detect and treat outliers if necessary
# We'll use boxplots to detect outliers visually
plt.figure(figsize=(15, 10))
sns.boxplot(data=df_clean)
plt.title("Boxplots to Detect Outliers")
plt.xticks(rotation=90)
plt.show()

# You can decide to remove outliers or treat them based on the boxplot results
# Example to remove outliers: df_clean = df_clean[(df_clean['column'] < upper_limit) & (df_clean['column'] > lower_limit)]

# 2. Statistical Summary

# Statistical summary for each variable
summary = df_clean.describe()
print("Statistical Summary:\n", summary)

# Highlight interesting findings
# For example:
# print("\nMean of 'LB':", df_clean['LB'].mean())
# print("Standard Deviation of 'MSTV':", df_clean['MSTV'].std())

# 3. Data Visualization

# Histograms for numerical variables
df_clean.hist(figsize=(15, 10), bins=20, edgecolor='black')
plt.suptitle("Histograms of Numerical Variables")
plt.show()

# Boxplots to visualize distributions
plt.figure(figsize=(15, 10))
sns.boxplot(data=df_clean)
plt.title("Boxplots of Numerical Variables")
plt.xticks(rotation=90)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_clean.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# Pair plot to explore relationships between variables
sns.pairplot(df_clean)
plt.suptitle("Pair Plot of the Dataset", y=1.02)
plt.show()

# Violin plot to visualize distributions and density
plt.figure(figsize=(15, 10))
sns.violinplot(data=df_clean)
plt.title("Violin Plot of Numerical Variables")
plt.xticks(rotation=90)
plt.show()

# 4. Pattern Recognition and Insights

# Identify correlations between variables
corr_matrix = df_clean.corr()
high_corr = corr_matrix[corr_matrix > 0.7]
print("\nHighly Correlated Variables:\n", high_corr)

# Look for trends or patterns
# Example: Is there any trend between Fetal Movements (FM) and Baseline FHR (LB)?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='FM', y='LB', data=df_clean)
plt.title("Scatter Plot between FM and LB")
plt.show()

# 5. Conclusion

# Summarize the key insights and patterns
print("Summary of Insights:")
print("- The correlation heatmap indicates strong relationships between certain variables such as [example variables].")
print("- [Additional insights based on your observations].")
print("- Consider exploring [suggestions for further analysis or recommendations].")

# Save the cleaned dataset for further analysis if needed
df_clean.to_csv('cleaned_cardiographic.csv', index=False)
```
